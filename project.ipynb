```python
import networkx as nx

f = open("filtered_friendship_network.txt","r")
G = nx.parse_edgelist(f,delimiter = ",",nodetype = int)
f.close()
cen = nx.closeness_centrality(G)
highest_cen_node =[]
sorted_cen = dict(sorted(cen.items(), key=lambda item: item[1], reverse = True))
print("The following is the the most central nodes(closeness):")
percent = len(sorted_cen) * 0.1
for i in range(round(percent)):
    highest_cen_node.append(list(sorted_cen.keys())[i])
print(highest_cen_node)
#use highest_cen_node variable, it is the array contains the top10% nodes
#same for all blocks down
```
